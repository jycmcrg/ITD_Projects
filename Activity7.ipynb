<h1 style=background-color:#D9EAFD>FILTERING DATA</h1>

```python
# FILTER OF REGION
# 50474 rows
filter1 = df[df['region_txt'] == 'Middle East & North Africa']
filter1

# FILTER OF ATTACK/CRIME TYPE
# 30908 rows
filter2 = filter1[filter1['attacktype1_txt'] == 'Bombing/Explosion']
filter2

# FILTER OF NKILL
# 1893 rows
filter3 = filter2[filter2['nkill'] >= 10]
filter3

# FILTER OF IYEAR
# 1737 rows
filter4 = filter3[filter3['iyear'] >= 2000]
filter4
```

<H1 style=background-color:#D9EAFD>DATA AGGREGATION</H1>

```python
# GROUPBY OF LATITUDE AND LONGITUDE
ME_terror = filter4.groupby(['longitude', 'latitude']).size().reset_index(name = 'incident_count')
ME_terror
```

<H1 style=background-color:#D9EAFD>Visualization of  GeoSpatial using FOLIUM</H1>

```python
import folium
from folium.plugins import HeatMap
```

<H1 style=background-color:LIGHTBLUE><b><i>Regional Analysis</H1>
<H2 style=background-color:#d7f9f8>
<b>Region:</b><i> Middle East & North Africa<br>
<b>Attack Type:</b><i> Bombing/Explosion<br>
<b>Casualties:</b><i> 10 or more<br>
<b>Year:</b><i> 2000 and later</H2>

```python
# CONVERT TO A LIST
heat_df = ME_terror[['longitude', 'latitude', 'incident_count']].values.tolist()
heat_df
```

```python
# CREATE OWN MAP
# Zoom Start (1,2,3,4)
my_map = folium.Map(location = [30, 40], zoom_start = 4)
my_map
```

```python
# ADD TO HEATMAP
HeatMap(heat_df).add_to(my_map)
```

```python
# CALL THE MAP
print(f"Filtered Dataset: {len(filter4)} Rows")
my_map
```

```python
# MAKE A HTML
my_map.save('Middle_East_&_North_Africa_Regional.html')
```

<H1 style=background-color:LIGHTBLUE><b><i>Regional Analysis #2</H1>
<H2 style=background-color:#d7f9f8>
<b>Region:</b><i> Southeast Asia<br>
<b>Year:</b><i> Before 2000<br>
<b>Attack Type:</b><i> Hostage Taking (Kidnapping)<br>
<b>Kill:</b><i> Less than 1</H2>

```python
regional = df[df['region_txt'] == 'Southeast Asia']
regional1 = regional[regional['iyear'] < 2000]
regional2 = regional1[regional1['attacktype1_txt'] == 'Hostage Taking (Kidnapping)']
regional3 = regional2[regional2['nkill'] < 1]
print(f"Filtered Dataset: {len(regional3)} Rows")

agg = regional3.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

regional_map = agg[['latitude', 'longitude', 'incident_count']].values.tolist()

map_reg = folium.Map(location=[12, 122], zoom_start=4)
HeatMap(regional_map).add_to(map_reg)
map_reg.save('Southeast_Asia_Regional_Map.html')
map_reg
```

<H1 style=background-color:LIGHTBLUE><b><i>Crime/Attack Analysis #3</H1>
<H2 style=background-color:#d7f9f8>
<b>Attack Type:</b><i> Facility/Infrastructure Attack<br>
<b>Year:</b><i> Before 2000<br>
<b>Kill:</b><i> Less than 10</H2>

```python
crime = df[df['attacktype1_txt'] == 'Facility/Infrastructure Attack']
crime2 = crime[crime['iyear'] < 2000]
crime3 = crime2[crime2['nkill'] < 10]
print(f"Filtered Dataset: {len(crime3)} Rows")

crime_agg = crime.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

crime_map = crime_agg[['latitude', 'longitude', 'incident_count']].values.tolist()

map_crime = folium.Map(location=[20, 5], zoom_start=2)
HeatMap(crime_map).add_to(map_crime)
map_crime.save('Facility_Infrastructure_Attack_Crime_Map.html')
map_crime
```

<H1 style=background-color:LIGHTBLUE><b><i>Nationality Analysis #4</H1>
<H2 style=background-color:#d7f9f8>
<b>Region:</b><i> North America<br>
<b>Attack Type:</b><i> Armed Assault<br>
<b>Nationality:</b><i> United States</h2>

```python
nationality = df[df['region_txt'].str.contains('North America')]
nationality1 = nationality[nationality['attacktype1_txt'] == 'Armed Assault']
nationality2 = nationality1[nationality1['natlty1_txt'] == 'United States']
print(f"Filtered Dataset: {len(nationality2)} Rows")

nationality_agg = nationality2.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

nationality_map = nationality_agg[['latitude', 'longitude', 'incident_count']].values.tolist()

map_nationality = folium.Map(location=[30, -96], zoom_start=4)
HeatMap(nationality_map).add_to(map_nationality)
map_nationality.save('United_States_Nationality_Map.html')
map_nationality
```

<H1 style=background-color:LIGHTBLUE><b><i>Casualty Analysis #5</H1>
<H2 style=background-color:#d7f9f8>
<b>Kill:</b><i> Less than 100<br>
<b>Attack Type:</b><i> Hijacking</h2>

```python
casualty = df[df['nkill'] < 100]
casualty1 = casualty[casualty['attacktype1_txt'] == 'Hijacking']
print(f"Filtered Dataset: {len(casualty1)} Rows")

casualty_agg = casualty1.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

casualty_map = casualty_agg[['latitude', 'longitude', 'incident_count']].values.tolist()

map_casualty = folium.Map(location=[15, 13], zoom_start=2)
HeatMap(casualty_map).add_to(map_casualty)
map_casualty.save('One_Hundred_Casuality_Map.html')
map_casualty
```

<H1 style=background-color:LIGHTBLUE><b><i>Target Analysis #6</H1>
<H2 style=background-color:#d7f9f8>
<b>Region:</b><i> South America<br>
<b>Year:</b><i> 1970 and later<br>
<b>Target Type:</b><i> Government (General)<br>
<b>Attack Type:</b><i> Assassination</H2>

```python
target = df[df['region_txt'].str.contains('South America')]
target1 = target[target['iyear'] >= 1970]
target2 = target1[target1['targtype1_txt'] == 'Government (General)']
target3 = target2[target2['attacktype1_txt'] == 'Assassination']
print(f"Filtered dataset size: {len(target3)} rows")

target_agg = target.groupby(['latitude', 'longitude']).size().reset_index(name='incident_count')

target_map = target_agg[['latitude', 'longitude', 'incident_count']].values.tolist()

map_target = folium.Map(location=[-20, -60], zoom_start=3)
HeatMap(target_map).add_to(map_target)
map_target.save('South_America_Target_Map.html')
map_target
```
