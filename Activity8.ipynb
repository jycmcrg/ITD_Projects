<H1 style=background-color:LIGHTBLUE><b> Primary Type #1:</b><i> Motor Vehicle Theft</H1>

```python
type1 = crimes[crimes['Primary Type'] == 'MOTOR VEHICLE THEFT']
print(f"Filtered Dataset: {len(type1)} Rows")

agg_type1 = type1.groupby(['Latitude', 'Longitude']).size().reset_index(name = 'incident_count')
agg_type1['normalized_weight'] = (agg_type1['incident_count'] - agg_type1['incident_count'].min()) / \
                                   (agg_type1['incident_count'].max() - agg_type1['incident_count'].min())

import folium
from folium.plugins import HeatMap

heat_vehivletheft = agg_type1[['Latitude', 'Longitude', 'normalized_weight']].values.tolist()

vehivletheft_map = folium.Map(location = [41.87811, -87.62980], zoom_start = 10)

HeatMap(heat_vehivletheft).add_to(vehivletheft_map)
vehivletheft_map.save('primarytype_motor_vehicle_theft.html')
vehivletheft_map
```
