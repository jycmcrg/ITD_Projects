```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
dengue = pd.read_csv('datasets\\dengue.csv')
dengue
```

<h1>Check Null</h1>

```python
dengue.isnull().sum()
```

```python
dengue.dtypes
```

<h1>Change to Category</h1>

```python
dengue['Month'] = dengue['Month'].astype('category')
dengue['Region'] = dengue['Region'].astype('category')

dengue.dtypes
```

<h1 style=background-color:yellow>Q1. Which Region have the Highest total dengue cases?</h1>
<h2>Insight #1: <b>REGION IV-A</b> have the highest total dengue cases in Region.</h2>
<h1 style=background-color:yellow>Q2. Which Region have the Lowest total dengue cases?</h1>
<h2>Insight #2: <b>BARMM</b> have the lowest total dengue cases in Region.</h2>

```python
region_cases = dengue.groupby(["Region"])['Dengue_Cases'].sum().reset_index()
region_cases
```

```python
region_cases = dengue.groupby(["Region"])['Dengue_Cases'].sum().reset_index()
sns.barplot(data=region_cases, x="Dengue_Cases", y="Region")
plt.title('REGION that have the HIGHEST and LOWEST Total Dengue Cases')
plt.show()
```

<h1 style=background-color:yellow>Q16. Do the Denge Deaths consistently peak in the same months every year?</h1>

<h1 style=background-color:yellow>Q17. Which Month had the Highest Number of Dengue Deaths in 2016?</h1>
<h2>Insight #17: The Highest number of Dengue Deaths in 2016 is the month of <b>OCTOBER</b>. While <b>APRIL</b> has the lowest number of dengue deaths in 2016.</h2>

```python
dengue_2016 = dengue[dengue['Year'] == 2016]
dengue_2016
```

```python
monthly_deaths_2016 = dengue_2016.groupby('Month')['Dengue_Deaths'].sum()
monthly_deaths_2016
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

monthly_deaths_2016 = monthly_deaths_2016.reindex(month_order)
sns.barplot(x=monthly_deaths_2016.index, y=monthly_deaths_2016.values)
plt.title('Dengue Deaths in 2016')
plt.gcf().set_size_inches(20, 8)
plt.show()
```
